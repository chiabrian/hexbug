```python
import pyvisa as visa
import time
import matplotlib.pyplot as plt
import copy
import pickle
%matplotlib tk
# # Replace with your instrument's actual VISA address
SCOPE_VISA_ADDRESS = 'USB0::0x0957::0x179A::MY51137476::INSTR' 

# Open the connection
rm = visa.ResourceManager()
scope = rm.open_resource(SCOPE_VISA_ADDRESS)

# print(rm.list_resources())

# Example commands
print(f"Connected to: {scope.query('*IDN?')}") # Query ID
# scope.write(':RUN') # Start the oscilloscope
# time.sleep(1)
# waveform_data = scope.query_binary_values(':WAVeform:DATA?', datatype='B') # Acquire waveform data

# # Close the connection
# scope.close()
```

```python
up = scope.query_binary_values(':WAVeform:DATA?', datatype='B', header_fmt='ieee')
```

```python
down = scope.query_binary_values(':WAVeform:DATA?', datatype='B', header_fmt='ieee')
```

```python
left = scope.query_binary_values(':WAVeform:DATA?', datatype='B', header_fmt='ieee')
```

```python
right = scope.query_binary_values(':WAVeform:DATA?', datatype='B', header_fmt='ieee')
```
